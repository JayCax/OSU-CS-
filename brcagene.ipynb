```python
gene = open("brca1.txt")
filecontents = gene.read()
print(filecontents)
```

```python
def replace_all(text, dic):
    for i, j in dic.items():
        text = text.replace(i, j)
    return text

mydna = filecontents
reps = {'A' : 'U' , 'T' : 'A' , 'G' : 'Y' , 'C' : 'G'}

myintrna = replace_all (mydna, reps)
print(myintrna)
```

```python
Acount = filecontents.count("A")
Tcount = filecontents.count("T")
Gcount = filecontents.count("G")
Ccount = filecontents.count("C")

print("The total number of Adenosines = " + str(Acount))
print("the total number of thymines = " + str(Tcount))
print("the total number of guanosines =" + str(Gcount))
print("the total number of cytosines = " + str(Ccount))
```

```python
myrna = myintrna
reps2 = {'Y' : 'C'}

myrna = replace_all (myintrna, reps2)
print(myrna)
```

```python
rnalength = len(myrna)
rnaAcount = myrna.count("A")
rnaUcount = myrna.count("U")
rnaGcount = myrna.count("G")
rnaCcount = myrna.count("C")

print("The toal product of transcription length = " + str(rnalength))
print("The total amount of RNA Adenosines = " + str(rnaAcount))
print("The total amount of RNA Uracils = " + str(rnaUcount))
print("The total amount of RNA Guanosines = " + str(rnaGcount))
print("The total amount of RNA Cytosines = " + str(rnaCcount))

GCrna = (rnaGcount + rnaCcount) / rnalength 
print("The GC ratio of the transcribed RNA product = " + str(GCrna))

AUrna = (rnaAcount + rnaUcount) / rnalength
print("The AU ratio of the transcribed RNA product = " + str(AUrna))

GCrna + AUrna
```

```python
ORF = myrna[125:17000]
print(ORF)
```

```python
ORF = ORF.replace("\n", "")
print(ORF)
```

```python
threelettercodon = ORF
[threelettercodon[i:i+3] for i in range(0, len(threelettercodon), 3)]
```

```python
aminoa = [threelettercodon[i:i+3] for i in range(0, len(threelettercodon), 3)]

for n, i in enumerate(aminoa):
    
    if i in ('CGA' , 'CGG' , 'CGC' , 'CGU' , 'AGG' , 'AGA'):
        aminoa[n] = 'arg'
        
    if i in ('CUU' , 'CUC' , 'CUA' , 'CUG' , 'UUA' , 'UUG'):
        aminoa[n] = 'leu'
        
    if i in ('UCU' , 'UCC' , 'UCA' , 'UCG' , 'AGU' , 'AGC'):
        aminoa[n] = 'ser'    
    
    if i == 'AUG':
        aminoa[n] = 'met'
        
    if i in ('GGA', 'GGU' , 'GGC' , 'GGG'):
        aminoa[n] = 'gly'
        
    if i in ('AAA' , 'AAG'):
        aminoa[n] = 'lys'
        
    if i in ('AUU' , 'AUC' , 'AUA'):
        aminoa[n] = 'ile'
        
    if i in ('GUU' , 'GUG' , 'GUA' , 'GUC'):
        aminoa[n] = 'val'
        
    if i in ('UUU' , 'UUC'):
        aminoa[n] = 'phe'
        
    if i in ('UGU' , 'UGC'):
        aminoa[n] = 'cys'
        
    if i in ('GCU' , 'GCA' , 'GCG' , 'GCC'):
         aminoa[n] = 'ala'
        
    if i in ('CCU' , 'CCC' , 'CCA' , 'CCG'):
        aminoa[n] = 'pro'
        
    if i in ('ACA' , 'ACU' , 'ACC' , 'ACG'):
        aminoa[n] = 'thr'
        
    if i in ('UAU', 'UAC'):
        aminoa[n] = 'tyr'
        
    if i in ('UGG'):
        aminoa[n] = 'trp'
        
    if i in ('CAA' , 'CAG'):
        aminoa[n] = 'gln'
        
    if i in ('GAA' , 'GAG'):
        aminoa[n] = 'glu'
        
    if i in ('GAC' , 'GAU'):
        aminoa[n] = 'asp'
        
    if i in ('AAU' , 'AAC'):
        aminoa[n] = 'asg'
        
    if i in ('CAU' , 'CAC'):
        aminoa[n] = 'his'
        
    if i in ('UAA' , 'UAG' , 'UGA'):
        aminoa[n] = 'STP'
        
print(aminoa)

```

```python
ala = aminoa.count('ala')
print("The total Alanine Count is:" + str(ala))

arg = aminoa.count('arg')
print("The total Argenine Count is:" + str(arg))

lys = aminoa.count('lys')
print("The total Lysine Count is:" + str(lys))

met = aminoa.count('met')
print("The total Methionine Count is:" + str(met))

STP = aminoa.count('STP')
print("The total STOP CODON Count is:" + str(STP))

```
